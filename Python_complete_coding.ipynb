```python
## EXample 1 temperature Conversion
def convert_temperature(temp,unit):
    """This function converts temperature between celsius and forenheit"""
    if unit=='C':
        return temp * 9/5 + 32 # celsius to forenheit
    elif unit=="F":
        return (temp-32)*5/9  ## Fahrenheit to celcius 
    else:
        return None

print(convert_temperature(25,'C'))
print(convert_temperature(77,'F'))

    
```

```python
#example 2 password checker
def is_strong_password(password):
    """this password checks if the password is strong or not"""
    if len(password)<8:
        
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False

    return True

## Calling the func
print(is_strong_password("WeakPWd"))
print(is_strong_password("Str0ngPWD!"))
        
```

```python
# EXample 3 : calculate the total Cost of item in a shopping cart
def calculate_total_cost(cart):
    total_cost=0
    for item in cart:
        total_cost+=item['price']* item['quantity']
    return total_cost    

# ex >> cart data
cart=[
    {'name':'Apple','price':0.5,'quantity':4},
    {'name':'Banana','price':0.3,'quantity':6},
    {'name':'Orange','price':0.7,'quantity':3}
]

## calling the fun
total_cost=calculate_total_cost(cart)
print(total_cost)
```

```python
## EX >> 4 check if a string IS Palindrome
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]
print(is_palindrome("A man a plan  a canal Panama"))
print(is_palindrome("Hello"))
    
```

```python
# ex 5 calcluate the factorials of a number using recursion
def factorial(n):
    if n==0:
        return 1 
    else:
        return n*factorial(n-1)

print(factorial(6))
```

```python
## try,except,else and finaaly
try:
    num=int(input("enter a number"))
    result=10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("enter denominator greater tha zer0!")
except Exception as ex:
    print(ex)
else:
    print(f"the result is {result}")
finally:
    
    print("Exceution Complete.")
```
