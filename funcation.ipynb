```python
def create_triangle(row_num):
    k = row_num - 1  # spaces before stars

    for i in range(row_num):
        # print spaces
        for j in range(k):
            print(" ", end="")
        
        # reduce spaces after each row
        k -= 1
        
        # print stars
        for j in range(i + 1):
            print("*", end="")
        
        print()  # move to the next line


create_triangle(3)
```
